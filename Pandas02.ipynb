```python
import pandas as pd
df = pd.read_csv("sales_data.csv")
df = df.head(20).set_index("invoice_no")
df = df.rename_axis(None)
df
```

```python
df.insert(4,"subtotal",df["quantity"] * df["price"])
df
```

```python
df.insert(5, "discount", df.apply(lambda row: round(row["subtotal"] * 0.05 if row["quantity"] > 3 else 0, 2), axis=1))
df
```

```python
df.insert(6,"total",df["subtotal"] - df["discount"])
df
```

```python
df1 = df[df["category"] == "Shoes"]
df1
```

```python
df2 = df[(df["shopping_mall"] == "Forum Istanbul") | (df["shopping_mall"] == "Zorlu Center")]
df2
```

```python
df3 = df[(df["shopping_mall"] == "Kanyon") & (df["category"] == 'Clothing')]
df3
```

```python
print(f"Total Penjualan Terendah : {round(min(df['total']),2)}")
print(f"Total Rata - Rata Penjualan : {round(df['total'].mean(),2)}")
print(f"Total Penjualan Tertinngi : {round(max(df['total']),2)}")
```

```python
total_2021 = df[df['invoice_date'].str[-4:] == '2021']['total'].sum()
total_2022 = df[df['invoice_date'].str[-4:] == '2022']['total'].sum()
total_2023 = df[df['invoice_date'].str[-4:] == '2023']['total'].sum()
print(f"Total Penjualan Tahun 2021: {total_2021}")
print(f"Total Penjualan Tahun 2022: {total_2022}")
print(f"Total Penjualan Tahun 2023: {total_2023:.2f}")
```
