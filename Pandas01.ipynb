```python
df2.insert(2,"coverage",65 - df2["age"])
df2
```

```python
df2.insert(len(df2.columns),"new_charges",(df2["charges"]*(0.05+df2["children"]*0.01))+df2["charges"])
df2.loc[df2["charges"]>5000,"new_charges"] = (df2["charges"]*(0.1+df2["children"]*0.01))+df2["charges"]
df2
```

```python
df2.insert(len(df2.columns),"Percent_increment",(0.05+(df2["children"]*0.01))*100)
df2.loc[df2["charges"]>5000,"Percent_increment"]=(0.1+(df2["children"]*0.01))*100
df2
```
